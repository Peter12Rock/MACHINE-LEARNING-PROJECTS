# WEB SCRAPING FOR SOME COMPANY STOCK

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://stockanalysis.com/stocks/screener/'

page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
# Locate and extract all table in the web page

soup.find_all('table')
```

```python
#Select specific table of interest

table = soup.find('table', class_ = "symbol-table svelte-1ie318d")

print(table)
```

```python
#Select the headings row

stock_heading = soup.find_all('th')

stock_heading
```

```python
#Create a list for the heading row

Allstock_heading = [heading.text for heading in stock_heading]

print(Allstock_heading)
```

```python
import pandas as pd
```

```python
#turn the heading into a panda Dataframe

df = pd.DataFrame(columns = Allstock_heading) 

df
```

```python
#select the other rows and assign a variable to them

column_data = table.find_all('tr')
```

```python
#Loop through the rows and create a lists 

for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    #Turn the list into a dataframe
    lenght = len(df)
    df.loc[lenght] = individual_row_data
```
